```python
sns.set_theme(style="white", context="talk")
import pandas as pd
data = pd.read_csv('data.csv', encoding='utf-8')
database = data[0:6]
dataadat = data[6:]
print(data)
depthdata = data[data['Category']=='Depth']
Gatedata = data[data['Category']=='Gatenumber']
# print(database)
print(depthdata)
print(Gatedata)

f, (ax1, ax2) = plt.subplots(2, 1,figsize=(20,15), sharex=False)

sns.barplot(data=depthdata,x="hue",y="value",hue="Baseline", palette="deep", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Depth-analysis")
ax1.set_xlabel("Benchmarks")

# Center the data to make it diverging
sns.barplot(data=Gatedata,x="hue",y="value",hue="Baseline", palette="deep", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Gate-number-analysis")
ax2.set_xlabel("Benchmarks")

sns.despine(bottom=True)
plt.setp(f.axes)
plt.tight_layout(h_pad=2)
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('compilation-4.csv', encoding='utf-8')
from matplotlib import rcParams
c = sns.color_palette("deep")
c2 = sns.color_palette("Paired")
c3 = sns.color_palette("Set2")
rcParams['figure.figsize'] = 15,8
sns.set(font_scale = 2.5)
g = sns.barplot(data = data, x="Benchmark", y = "Compilation_time",hue = "type",palette=c2)
g.set_yscale("log")
g.set_ylabel("Normalized depth")
g.set_xlabel("Benchmarks")
figure = g.get_figure()    
figure.savefig('bar.pdf', dpi=400)
```
